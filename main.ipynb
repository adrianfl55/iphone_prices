```python
import chardet as cd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Opción 1 - Creamos el Dataframe a mano

df = pd.DataFrame({
    "dispositivo": ["iPhone 8", "iPhone X", "iPhone XR", "iPhone XS", "iPhone XS Max", "iPhone 11", "iPhone 11 Pro", "iPhone 11 Pro Max", "iPhone 12", "iphone 12 Pro", "iPhone 12 Pro Max", "iPhone 13", "iPhone 13 Pro", "iPhone 13 Pro Max", "iPhone 14", "iPhone 14 Pro", "iPhone 14 Pro Max","iPhone 15", "iPhone 15 Pro", "iPhone 15 Pro Max", "iPhone 16", "iPhone 16 Pro", "iPhone 16 Pro Max"],
    "lanzamiento": [2017, 2017, 2018, 2018, 2018, 2019, 2019, 2019, 2020, 2020, 2020, 2021, 2021, 2021, 2022, 2022, 2022, 2023, 2023, 2023, 2024, 2024, 2024],
    "precio_dolar": [699, 999, 749, 999, 1099, 699, 999, 1099, 799, 999, 1099, 799, 999, 1099, 799, 999, 1099, 799, 999, 1199, 799, 999, 1199],
    "precio_euro": [809, 1159, 859, 1159, 1259, 809, 1159, 1259, 909, 1159, 1259, 909, 1159, 1259, 1009, 1319, 1469, 959, 1219, 1469, 959, 1219, 1469],
    "categoria": ["Base", "Pro", "Base", "Pro", "Pro Max", "Base", "Pro", "Pro Max", "Base", "Pro", "Pro Max", "Base", "Pro", "Pro Max", "Base", "Pro", "Pro Max", "Base", "Pro", "Pro Max", "Base", "Pro", "Pro Max"]
})

print(df)
```

```python
# Opción 2 - Creamos el Dataframe a partir de un archivo CSV

# Averiguamos primero la codificación del CSV
with open('precios.csv', 'rb') as f:
    result = cd.detect(f.read())

# Creamos el dataframe con la codificación correcta
df = pd.read_csv('precios.csv', delimiter = ';', index_col = False, encoding = result['encoding'])

df['categoria'] = df['categoria'].str.strip()

print(df)
```

```python
# Filtramos el DataFrame para seleccionar solo los dispositivos de la categoria "Base"
df_base = df[df['categoria'] == 'Base']

# Seleccionamos las columnas a representar
x = df_base['lanzamiento']
y1 = df_base['precio_euro'] 
y2 = df_base['precio_dolar']

# Creamos el gráfico
plt.plot(x, y1, color = "red", marker = "o", linestyle = "-", linewidth = 2, markersize = 5)
plt.plot(x, y2, color = "blue", marker = "o", linestyle = "-", linewidth = 2, markersize = 5)

# Añadimos etiquetas a los ejes
plt.xlabel('Año de lanzamiento')
plt.ylabel('Precio')

# Ponemos el eje X como tipo entero para que no haya fechas en .5
plt.xticks(x.astype(int))

# Añadimos un título al gráfico
plt.title('Evolución del precio de los iPhone Base', color = '#381f22')

# Añadimos leyenda
plt.legend(['Precio en euros', 'Precio en dólares'])

# Mostramos el gráfico
plt.show()
```
